```python
def create_contract(address,abi):
    return w3.eth.contract(
        address = w3.toChecksumAddress(address),
        abi= json.load(open(abi,'r'))
    )

def readValue(fn_call, block_identifier='latest', default_value=None,debug=False):
    try:
        return fn_call.call(block_identifier = block_identifier)
    except Exception as E:
        if  debug:
            print(E)
        return default_value

def getEvents(abi):
    abi=json.loads(abi)
    events=[]

    for i in abi:
        if i.get('type') == 'event' :
            events.append(i)
    return events

def getFunctions(abi):
    abi=json.loads(abi)
    functions=[]

    for i in abi:
        if i.get('stateMutability') == 'view':
            functions.append(i)
    return functions
```
