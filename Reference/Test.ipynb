```python
import json
from web3 import Web3
from collections import defaultdict
from web3.middleware import geth_poa_middleware
```

```python
w3 = Web3(Web3.HTTPProvider('https://mainnet.infura.io/v3/0290160ac7b5492d89913bba7746b772'))

w3.middleware_onion.inject(geth_poa_middleware,layer=0)
print(f'Connected: {w3.isConnected()}')
```

```python
def create_contract(address,abi):
    return w3.eth.contract(
        address = w3.toChecksumAddress(address),
        abi= json.load(open(abi,'r'))
    )

def readValue(fn_call, block_identifier='latest', default_value=None,debug=False):
    try:
        return fn_call.call(block_identifier = block_identifier)
    except Exception as E:
        if  debug:
            print(E)
        return default_value

def getEvents(abi):
    abi=json.loads(abi)
    events=[]

    for i in abi:
        if i.get('type') == 'event' :
            events.append(i)
    return events

def getFunctions(abi):
    abi=json.loads(abi)
    functions=[]

    for i in abi:
        if i.get('stateMutability') == 'view':
            functions.append(i)
    return functions
```

```python
gauge_controller_contract = create_contract(GAUGE_CONTROLLER,'./abi/gauge_controller.json')
GAUGES = []
idx = 0
while(True):
    gauge = readValue(gauge_controller_contract.functions.gauges(idx),default_value = "0x0000000000000000000000000000000000000000")
    idx = idx + 1
    if(gauge == "0x0000000000000000000000000000000000000000"):
        break
    GAUGES.append(gauge)

print(GAUGES)
```

```python
def getGaugeData(gaugeAddress):

    gauge_contract = create_contract(gaugeAddress,'./abi/gauge.json')
    name = readValue(gauge_contract.functions.name(),default_value = "")
    symbol = readValue(gauge_contract.functions.symbol(),default_value = "")
    decimals = readValue(gauge_contract.functions.decimals(),default_value = 0)
    lp_token = readValue(gauge_contract.functions.lp_token(),default_value = "")
    total_supply = readValue(gauge_contract.functions.totalSupply(),default_value = 0)
    working_supply = readValue(gauge_contract.functions.working_supply(),default_value = 0)
    inflation_rate = readValue(gauge_contract.functions.inflation_rate(),default_value = 0)
    reward_count = readValue(gauge_contract.functions.reward_count(),default_value = 0)
    reward_tokens = []
    reward_data = []
    
    print(f'Gauge Address: {gaugeAddress}\nName: {name}\nSymbol: {symbol}\nDecimals: {decimals}\nLpToken: {lp_token}\nTotal Supply: {total_supply}\nWorking Supply: {working_supply}\nInflastion Rate: {inflation_rate}\nReward Count: {reward_count}')

    i = 0
    while (True):
        reward_token = readValue(gauge_contract.functions.reward_tokens(i),default_value = "0x0000000000000000000000000000000000000000")
        i = i + 1
        if (reward_token == "0x0000000000000000000000000000000000000000"):
            break
        reward_token_data = readValue(gauge_contract.functions.reward_data(reward_token),default_value = {})
        print(f'Reward Token: {reward_token} --> \nToken Data: {reward_token_data}')
    print("\n---------------------------------------------------------------------------------------")


```
